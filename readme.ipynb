# Краткое содержание статьи "An artificial intelligence model for heart disease detection using machine learning algorithms" ("Модель искусственного интеллекта построенная с использованием алгоритмов машинного обучения для обнаружения заболеваний сердца").
## Авторы: Victor Chang, Vallabhanent Rupa Bhavani, Ariel Qianwen Xu, MA Hossain
## Healthcare Analytics, 2022, https://doi.org/10.1016/j.health.2022.100016

Авторы статьи демонстрируют как методы машинного обучения (МО) могут помочь в предсказании риска развития сердечно-сосудистых заболеваний у людей. Целью исследования является выявление сердечно-сосудистых заболеваний с использованием языка программирования Python. Конечным итогом данного проекта является готовое приложение для оценки рисков заболеваний. Основной текст статьи посвящен особенностям языка Python и обоснованием его выбора для разработки готового приложения, а также описанию нескольких используемых для проекта алгоритмов машинного обучения.
В статье описаны основные этапы разработки приложения: сбор данных для последующего анализа из баз данных, обработка датасета (в том числе работа с категориальными переменными, например преобразование категориальных признаков), тестирование нескольких алгоритмов МО для выполнения классификации. 

Авторы статьи в качестве языка программирования используют Python, а также используют такие пакеты и библиотеки, как Pandas, Matplotlib, IPython, Numpy, Python, SciPy и многие другие.

Используемый в статье набор данных состоит из 14 признаков, являющимися результатами медицинских тестов, проведённых примерно на 100 людях. Датасет состоит из следующих признаков: 

1.	Age (возраст);
2.	Sex (пол);
3.	Resting_blood_pressure (артериальное давление в состоянии покоя);
4.	Serum_cholestoral (уровень холестерина в крови mg/dl);
5.	Fasting_blood_sugar (анализ глюкозы в крови натощак (> 120 mg/dl));
6.	Maximum_heart_rate_achieved (максимальная частота сердечных сокращений);
7.	Exercise_induced_angina (стенокардия при нагрузке),
8.	Oldpeak (депрессия ST-сегмента);
9.	Slope (наклон пикового сегмента ST при физической нагрузке);
10.	Number_of_major_vessels (количество крупных сосудов (0-3));
11.	Resting_electrocardiographic_results (результаты электрокардиографических исследований в покое 0, 1, 2);
12.	Thal (3 = норма; 6 = фиксированный дефект; 7 = обратимый дефект);
13.	Chest_bin (боль в груди от 1 до 4);
14.	 Предиктор

Наличие сердечно-сосудистого заболевания выражено категориальной переменной, в которой 0 — ложное значение (у пациента нет заболеваний сердца), 1 — истинное значение (у пациента болезнь сердца).

Для оценки связей между несколькими переменными использовалась матрица корреляций. Так, авторы обнаружили положительную корреляцию между предиктором и признаком «боль в груди», было высказано предположение, что интенсивность боли в груди прямо пропорциональна вероятности сердечных заболеваний.

Авторы применили несколько алгоритмов МО для классификации данных и выбрали на их взгляд наиболее оптимальный. Для оценки алгоритмов МО использовались такие метрики качества, как accuracy, confusion matrix, precision, recall и F1 values.

Авторы получили следующие значения метрики precision для используемых алгоритмов:

+ алгоритм K-Neighbors classifier показал 87%;
+ Support vector machines 83%, 
+ Decision tree classifier 79%;
+ Random forest classifier 84%.

Для дальнейшей разработки приложения авторы остановились на алгоритме случайного леса (Random forest classifier).
